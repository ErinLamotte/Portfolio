####Visualizing Outliers with Boxplots

```python
sns.boxplot(data['Quantity'])
```

There are obvious outliers which will need to be removed. It looks like a high proportion of data lies below 0. These might be discounts, coupons, or returns. They could also be data entry mistakes in the dataset. Without knowing more, we will keep these records in the dataset.

```python
sns.boxplot(data['Price'])
```

Obvious outliers in the dataset that will need to be removed

####Removing Outliers with the Inter-Quartile Range Method

```python
Q1_Quantity = data['Quantity'].quantile(0.25)
Q3_Quantity = data['Quantity'].quantile(0.75)
IQR_Quantity = Q3_Quantity - Q1_Quantity

Q1_Price = data['Price'].quantile(0.25)
Q3_Price = data['Price'].quantile(0.75)
IQR_Price = Q3_Price - Q1_Price

data_outliersremoved = data[(data['Quantity'] >= (Q1_Quantity - 1.5 * IQR_Quantity)) & (data['Quantity'] <= (Q3_Quantity + 1.5 * IQR_Quantity)) &
                            (data['Price'] >= (Q1_Price - 1.5 * IQR_Price)) & (data['Price'] <= (Q3_Price + 1.5 * IQR_Price))]
```

```python
data_outliersremoved.describe()
```

After removing the outliers, the distribution can be more easily analyzed.

####Visualizing the Distribution with Histograms

```python
x=data_outliersremoved[['Quantity']].to_numpy()

plt.hist(x)
plt.xlabel('Quantity')
plt.ylabel('Frequency')
plt.title('Histogram of Quantity')

plt.show()
```

The histogram is fairly normal. The most frequent quantity per item lies in the bin between 0 and 5.

```python
x=data_outliersremoved[['Price']].to_numpy()

plt.hist(x)
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.title('Histogram of Price')

plt.show()
```

The histogram for Price is more right skewed than the histogram for Quantity. The most frequent price per unit appears to be $1.
